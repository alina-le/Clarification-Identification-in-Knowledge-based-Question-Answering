# Prompt ChatGPT with CLAQUA data

**Prompting**

Guidelines for how to formulate a prompt taken from: https://learn.deeplearning.ai/chatgpt-prompt-eng/lesson/2/guidelines

Four different prompts were developed for the task of clarification identification on the CLAQUA corpus. <br>
Single-turn sample prompted with: Prompt0, Prompt1, Prompt2 <br>
Multi-turn sample prompted with: Prompt0, Prompt1, Prompt2, Prompt3

**Evaluation:**

The responses are evaluated:

  1) Annotated based on the label: "yes, clarification needed" corresponds to class label 1 and "no, clarification not needed" corresponds to class label 0. These label-based evaluations can be either correct or incorrect with respect to the gold label.

  2) For items which receive the correct label in the GPT-response, it is annotated whether the explanation for (no) clarification need is correct or incorrect.

  3) Each response is evaluated for whether it contains:
   - hallucination
   - omission
   - incoherence
   - focus-deviation
   
The annotation is binary: <br>
0 for no hallucination/omission/incoherence/focus-deviation in the response, <br>
1 for hallucination(s)/omission/incoherence/focus-deviation in the response

````python
model = 'gpt-3.5-turbo'
````

````python
# taken from https://learn.deeplearning.ai/chatgpt-prompt-eng/lesson/2/guidelines
def get_completion(prompt, model=model):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

## Evaluate single-turn test samples

````python
def get_claqua_item_concat(idx):
  concat_item = single_turn_test_sample.iloc[idx]['context'] + " " + single_turn_test_sample.iloc[idx]['entity1'] + " " + single_turn_test_sample.iloc[idx]['entity2']
  print(single_turn_test_sample.iloc[idx]['context'])
  print(single_turn_test_sample.iloc[idx]['entity1'])
  print(single_turn_test_sample.iloc[idx]['entity2'])
  print(single_turn_test_sample.iloc[idx]['label'])
  return concat_item
````

### Evaluate prompt0 response

````python
for idx in range(len(single_turn_test_sample)):

  if idx % 10 == 0:
    print(f"Prompting item {idx}/{len(single_turn_test_sample)}")

  claqua_prompt0 = f"""

  Does the following user question to a knowledge-based question answering system need a clarification request or not?
  Answer with 'yes' or 'no'.

  User question:
  ```{single_turn_test_sample.iloc[idx]['context']}```
  Entity1 in the knowledge base:
  ```{single_turn_test_sample.iloc[idx]['entity1']}```
  Entity2 in the knowledge base:
  ```{single_turn_test_sample.iloc[idx]['entity2']}```
  """
  prompt0_response = get_completion(claqua_prompt0)

  if prompt0_response.casefold() == "yes":  # clarification need
    single_turn_test_sample.loc[single_turn_test_sample.index[idx], 'prompt0_label'] = 1
  elif prompt0_response.casefold() == "no":  # no clarification need
    single_turn_test_sample.loc[single_turn_test_sample.index[idx], 'prompt0_label'] = 0
  else:
    raise ValueError(f'GPT response must be either "yes" or "no" but is {prompt0_response}')

  time.sleep(20)
````
